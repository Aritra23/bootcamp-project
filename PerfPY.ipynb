```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# File to Load (Remember to change these)
performance_states = "naep.csv"

# Read the City and Ride Data
perf_states_df = pd.read_csv(performance_states)

# Combine the data into a single dataset
#combined_data_df = pd.merge(city_data_df, ride_data_df, how='outer', on='city')
#combined_data_df

# Display the data table for preview
math_type = perf_states_df.loc[perf_states_df['TEST_SUBJECT'] == 'Mathematics']
math_yr4_type = math_type.loc[math_type['TEST_YEAR'] == 4]
math_yr4_type.to_csv("mathyr4.csv", encoding="utf-8", index=False)

math_yr8_type = math_type.loc[math_type['TEST_YEAR'] == 8]
math_yr8_type.to_csv("mathyr8.csv", encoding="utf-8", index=False)

read_type = perf_states_df.loc[perf_states_df['TEST_SUBJECT'] == 'Reading']
read_yr4_type = read_type.loc[read_type['TEST_YEAR'] == 4]
read_yr4_type.to_csv("readyr4.csv", encoding="utf-8", index=False)

read_yr8_type = read_type.loc[read_type['TEST_YEAR'] == 8]
read_yr8_type.to_csv("readyr8.csv", encoding="utf-8", index=False)
```

```python
math_ne = math_yr4.loc[math_yr4['REGION'] == 'Northeast']
math_ne_avg = math_ne.groupby('YEAR').agg({'AVG_SCORE':'mean'})

math_mw = math_yr4.loc[math_yr4['REGION'] == 'Midwest']
math_mw_avg = math_mw.groupby('YEAR').agg({'AVG_SCORE':'mean'})

math_s = math_yr4.loc[math_yr4['REGION'] == 'South']
math_s_avg = math_s.groupby('YEAR').agg({'AVG_SCORE':'mean'})

math_w = math_yr4.loc[math_yr4['REGION'] == 'West']
math_w_avg = math_w.groupby('YEAR').agg({'AVG_SCORE':'mean'})
```

```python
plt.plot(math_ne_avg.index, math_ne_avg['AVG_SCORE'], color='green', marker='o', linestyle='dashed', linewidth=2, markersize=8)
plt.plot(math_mw_avg.index, math_mw_avg['AVG_SCORE'], color='blue', marker='+', linestyle='dashed', linewidth=2, markersize=8)
plt.plot(math_s_avg.index, math_s_avg['AVG_SCORE'], color='red', marker='x', linestyle='dashed', linewidth=2, markersize=8)
plt.plot(math_w_avg.index, math_w_avg['AVG_SCORE'], color='black', marker='^', linestyle='dashed', linewidth=2, markersize=8)
plt.grid()
plt.title("Year vs. Avg Grade 4 Math Scores")
plt.xlabel("Year")
plt.ylabel("Avg Grade 4 Math Score")
plt.show()
```
