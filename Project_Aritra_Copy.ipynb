```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("data_sets/states.csv")
df
```

```python
update_df = df.dropna(subset = ['ENROLL'])
update_df
```

```python
update_df["Amount Spend"] = update_df["TOTAL_REVENUE"] - update_df["TOTAL_EXPENDITURE"]
update_df
```

```python
state_amount_final = update_df.groupby("STATE")["Amount Spend"].mean()
#print(state_amount_final)
positive_spend_state_count = (state_amount_final > 0).sum()
print("Total States with positive final result:" + str(positive_spend_state_count))
```

```python
update_df["REGION"] = np.nan

update_df.loc[update_df.STATE.isin(['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire',
                               'Rhode Island','Vermont','New Jersey', 'New York',
                               'Pennsylvania']), 'REGION'] = 'Northwest'

update_df.loc[update_df.STATE.isin(['Illinois', 'Indiana', 'Michigan', 'Ohio','Wisconsin','Iowa', 'Kansas',
                               'Minnesota', 'Missouri', 'Nebraska', 'North Dakota',
                               'South Dakota']), 'REGION'] = 'Midwest'

update_df.loc[update_df.STATE.isin(['Delaware', 'Florida', 'Georgia', 'Maryland', 'North Carolina',
                               'South Carolina', 'Virginia','District of Columbia','West Virginia',
                               'Alabama', 'Kentucky', 'Mississippi', 'Tennessee',
                               'Arkansas', 'Louisiana', 'Oklahoma', 'Texas']), 'REGION'] = 'South'

update_df.loc[update_df.STATE.isin(['Arizona','Colorado', 'Idaho', 'Montana', 'Nevada', 'New Mexico',
                               'Utah','Wyoming','Alaska', 'California', 'Hawaii', 'Oregon',
                               'Washington']), 'REGION'] = 'West'

update_df
```

```python
state_amount_final = update_df.groupby("YEAR")["Amount Spend"].mean()

print("Positive final results since 1993")
print((state_amount_final > 0).sum())

plt.figure(figsize = (15,6))
g = sns.barplot(x="YEAR", y="Amount Spend",data=update_df)
g.set_title("Final Result REVENUE X EXPENDITURE", fontsize=15)
g.set_xlabel("Years ", fontsize=12)
g.set_ylabel("Net Income", fontsize=12)

plt.show()
```

```python
#enroll_per_student = update_df.groupby(["STATE", "YEAR"])

update_df["Expenditure per Student"] = ((update_df["TOTAL_EXPENDITURE"]*1000) / (update_df["ENROLL"]))
update_df.head()
```

```python
fig = plt.figure(figsize = (15,20))
ax = fig.add_subplot(5, 1, 1)
g = sns.lineplot(x="YEAR", y="Expenditure per Student",data=update_df)
g.set_title("Expenditure on each enrolled student per year", fontsize=15)
ax.xaxis.set_major_locator(ticker.MultipleLocator(1.00))
ax.grid(b=True, which='major', color='b', linewidth=0.5)
ax.grid(b=True, which='minor', color='b', linewidth=0.5)
g.set_xlabel("Years ", fontsize=12)
g.set_ylabel("Expenditure per student enroll", fontsize=12)

plt.show()
```

```python
update_df["Revenue per Student"] = ((update_df["TOTAL_REVENUE"]*1000) / (update_df["ENROLL"]))
update_df.head()
```

```python
fig = plt.figure(figsize = (15,20))
ax = fig.add_subplot(5, 1, 1)
g = sns.lineplot(x="YEAR", y="Revenue per Student",data=update_df)
g.set_title("Revenue on each enrolled student per year", fontsize=15)
ax.xaxis.set_major_locator(ticker.MultipleLocator(1.00))
ax.grid(b=True, which='major', color='b', linewidth=0.5)
ax.grid(b=True, which='minor', color='b', linewidth=0.5)
g.set_xlabel("Years ", fontsize=12)
g.set_ylabel("Revenue per student enroll", fontsize=12)

plt.show()
```
